```python

def load_logs_to_dataframe(file_path: str) -> pd.DataFrame:
    """
    Load a JSON log file into a flattened pandas DataFrame.

    Parameters:
        file_path (str): Path to the JSON file.

    Returns:
        pd.DataFrame: Flattened DataFrame with timestamp converted to datetime.
    """
    # Load the JSON file
    df = pd.read_json(file_path)
    
    # Flatten the nested 'data' column
    df = pd.json_normalize(df.to_dict(orient="records"))
    
    # Clean column names
    df.columns = df.columns.str.replace("data.", "")
    
    # Convert timestamp column to datetime
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    
    return df
log_adv = load_logs_to_dataframe("../../../data/log.advanced.broken.json")
log_adv
```

```python
def check_trains_with_switches(df, connections):
    trains = []
    waiting_sensors = []
    errors = []
    active_switches = {}
    pending_invalid = []

    for idx, row in df.iterrows():
        sensor = row["sensor"]
        direction = row["direction"] if not pd.isna(row["direction"]) else None
        timestamp = row["timestamp"]

        # 0. Switch events
        if direction is None and sensor.startswith("SS"):
            waiting_sensors = [(ts, s) for ts, s in waiting_sensors if s != sensor]
            active_switches[sensor] = 1
            continue

        # 1. Starting points (WITH direction constraint)
        if sensor in ["AC1.1", "AC1.10"]:
            start_ok = True
            for train in trains:
                _, last_sensor, _ = train[-1]
                if sensor == "AC1.1" and last_sensor == "AC1.2":
                    start_ok = False
                if sensor == "AC1.10" and last_sensor == "AC1.9":
                    start_ok = False

            if start_ok:
                trains.append([(timestamp, sensor, direction)])
                continue

        # 2. Sensors without direction
        if direction is None:
            waiting_sensors.append((timestamp, sensor))
            active_switches[sensor] = 1
            continue

        # 3. Match to train
        matched_train = None
        corrected_direction = direction

        for train in trains:
            last_ts, last_sensor, last_dir = train[-1]
            options = connections.get(last_sensor, [])

            # ✅ FIRST: check direct match
            direct_match = any(
                next_sensor == sensor and (next_dir is None or next_dir == direction)
                for next_sensor, _, _, next_dir in options
            )

            # ✅ ONLY if no direct match → try inserting ONE valid intermediate
            if not direct_match:
                inserted = False

                for inter_sensor, _, _, _ in options:
                    # ❌ avoid loops / duplicates
                    if inter_sensor == last_sensor or inter_sensor == sensor:
                        continue

                    # ❌ avoid inserting something that is literally next in log
                    if idx + 1 < len(df):
                        if df.iloc[idx + 1]["sensor"] == inter_sensor:
                            continue

                    # check valid path: last → inter → current
                    next_options = connections.get(inter_sensor, [])
                    if not any(ns2 == sensor for ns2, _, _, _ in next_options):
                        continue

                    # ✅ VALID → insert ONCE
                    new_time = last_ts + pd.Timedelta(seconds=10)
                    new_dir = last_dir

                    train.append((new_time, inter_sensor, new_dir))
                    errors.append(
                        f"Inserted missing sensor {inter_sensor} between {last_sensor} and {sensor} at {new_time}"
                    )

                    inserted = True
                    break

                if inserted:
                    # refresh after insertion
                    last_ts, last_sensor, last_dir = train[-1]
                    options = connections.get(last_sensor, [])

            # 🔁 NORMAL MATCHING
            for next_sensor, conn_type, switch_name, next_dir in options:
                if next_sensor != sensor:
                    continue

                if next_dir is not None:
                    corrected_direction = next_dir

                if conn_type == "switch":
                    found_switch = False
                    for ws_idx, (ws_ts, ws_sensor) in enumerate(waiting_sensors):
                        if ws_sensor == switch_name:
                            train.append((ws_ts, ws_sensor, "switch"))
                            waiting_sensors.pop(ws_idx)
                            found_switch = True
                            break

                    if not found_switch:
                        train.append((timestamp, switch_name, "switch"))

                    matched_train = train
                    if switch_name in active_switches:
                        del active_switches[switch_name]
                    break
                else:
                    matched_train = train
                    break

            if matched_train:
                break

        # 4. Append or store pending
        if matched_train:
            matched_train.append((timestamp, sensor, corrected_direction))
        else:
            pending_invalid.append((sensor, direction, timestamp))

        # 5. Pending logic (unchanged)
        if pending_invalid:
            new_pending = []
            for p_sensor, p_dir, p_ts in pending_invalid:
                appended = False
                for future_idx in range(idx + 1, len(df)):
                    f_row = df.iloc[future_idx]
                    f_ts = f_row["timestamp"]

                    for train in trains:
                        _, last_sensor, last_dir = train[-1]
                        options = connections.get(last_sensor, [])
                        for next_sensor, _, _, next_dir in options:
                            if next_sensor == p_sensor:
                                final_dir = next_dir if next_dir is not None else p_dir
                                new_time = f_ts + pd.Timedelta(seconds=10)
                                train.append((new_time, p_sensor, final_dir))
                                appended = True
                                break
                        if appended:
                            break
                    if appended:
                        break
                if not appended:
                    new_pending.append((p_sensor, p_dir, p_ts))
            pending_invalid = new_pending

    # 6. Errors
    for p_sensor, p_dir, p_ts in pending_invalid:
        errors.append(f"Sensor {p_sensor} at {p_ts} could not be matched to any train")

    if waiting_sensors:
        ws_list = ", ".join([s for _, s in waiting_sensors])
        errors.append(f"Waiting sensors not appended: {ws_list}")

    # 7. Summary
    train_summaries = []
    for train_idx, train in enumerate(trains):
        sensors_passed = [s for _, s, _ in train]
        train_summaries.append((train_idx, sensors_passed))

    return trains, train_summaries, errors
# Example usage:
trains, train_summaries, errors = check_trains_with_switches(log_adv, connections)

print(f"Detected {len(trains)} trains in the log.\n")

for train_idx, sensors in train_summaries:
    print(f"Train {train_idx} passed sensors in order: {sensors}")

if errors:
    print("\nInvalid movements detected:")
    for e in errors:
        print(e)
else:
    print("\nAll movements follow the connections logic for all trains.")
```
